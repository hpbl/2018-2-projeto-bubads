# Research Guide


## Research Goals

We have to main research goals:
1. To map the current state of the art regarding iOS open-soource frameworks/libraries README's.
2. To provide insights to maintainers on how to structure their README.md files.

## Our own questions

- Does the order of the sections follow a common order?
- Are gifs and images used to contextualize concepts (not just the logo)?
- Contributing guidelines are provided on the README, or at another file?
- Usage guide is provided on the README, or at another file?
- Is the building status present?
- Which badges are used?
- Are HTML tags used?
- Which HTML tags are most used?
